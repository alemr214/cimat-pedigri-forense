# 01. Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read data

```python
db = pd.read_excel("../data/raw/archivo_combinado.xlsx")
```

```python
# Split compund columns
db[['FAM_Sample_info_id', 'FAM_Sample_info_rel']] = db['FAM_Sample_info'].str.split(' ', n = 1, expand = True)
db[['VIC_Sample_info_id', 'VIC_Sample_info_rel']] = db['VIC_Sample_info'].str.split(' ', n = 1, expand = True)
db.drop(['FAM_Sample_info', 'VIC_Sample_info'], axis = 1, inplace = True)
```

```python
db
```

```python
db.drop_duplicates(subset =  [])
```

## Explore data

```python
VICs = db["VIC_Sample_info_id"].unique()
print(f"Number of VICs id's: {len(VICs)}")
```

```python
VICs_counts = db["VIC_Sample_info_id"].value_counts()
VICs_counts.plot(kind = 'hist')

plt.ylabel('Frecuencia')
plt.xlabel('Frecuencia de VICs')
plt.title('Histogram de VIC_Sample_info_id')
```
