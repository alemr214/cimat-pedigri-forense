# Revisar valores repetidos

```python
a = db[["Shared_Markers", "k0", "k1", "k2", "Share_allele", "True_positive"]].copy()
a = a.groupby(["Shared_Markers", "k0", "k1", "k2", "Share_allele"]).size().reset_index()
```

```python
b = db[["Shared_Markers", "k0", "k1", "k2", "Share_allele", "True_positive"]].copy()
b = b.groupby(["Shared_Markers", "k0", "k1", "k2", "Share_allele"]).sum().reset_index()
```

```python
a
```

```python
b
```

```python
result = pd.merge(a, b, on=["Shared_Markers", "k0", "k1", "k2", "Share_allele"], how='inner')
result.rename(columns = {0: 'count', 'True_positive': 'positives'}, inplace= True)
result
```

```python
# Definir una función para crear la nueva columna
def calc_discrepancy(row):
  if row['positives'] == row['count']:
    return 0
  else:
    return row['positives']

# Aplicar la función a cada fila
result['discrepancy'] = result.apply(calc_discrepancy, axis=1)
result
```

```python
result["discrepancy"].hist()
plt.title("Cantidad de labels que difieren de la mayoría")
plt.xlabel("Cantidad de labels diferentes")
plt.ylabel("Frecuencia")
```

## Graficar el espacio

```python
db = db_bkup.copy()
db = db[db.Kinship == "Paternidad"]
```

```python
db["True_positive"] = np.where(db["True_positive"] == False, 0, 1)
```

```python
db.drop(db[db.k0 + db.k1 + db.k2 != db.Marcadores].index, inplace=True)
```

```python
db_tograph = db[["k1", "k2", "ValueLog10", "True_positive"]].copy()
```

```python
db_tograph = db_tograph.groupby(["k1", "k2", "ValueLog10"]).mean().reset_index()
db_tograph
```

```python
# Create a figure and a 3D axes object
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Create the 3D scatter plot
ax.scatter(db_tograph['k1'], 
           db_tograph['k2'], 
           db_tograph['ValueLog10'],
           c = db_tograph['True_positive'],
           cmap='coolwarm',
           marker='o')

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set a title for the plot
ax.set_title('Simple 3D Scatter Plot')

# Show the plot
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(db_tograph, x='k1', y='k2', z='ValueLog10',
              color='True_positive')
fig.update_traces(marker_size = 5)
fig.show()
```
